```python
# visualize conditions for which a fix-point exists or not for a linear function

import holoviews as hv
import numpy as np
hv.extension('bokeh')
import bokeh.palettes as bp  # for color palettes
```

```python
# Define line a*x+b
n = 201
a = 0.8  # slope
b = 2  # offset
x = np.linspace(0,25,n)
def fct(a,x,b):
    return a*(x-2)+b
y = fct(a,x,b)

n_a = 200 # number of slope changes
lower_slope_limit = 0.4
upper_slope_limit = 1.5

# let a vary between lower and upper_slope_limit
a_var = np.linspace(lower_slope_limit,upper_slope_limit,n_a)
x_range = np.linspace(5,3,n_a)

y_list = []
for i in range(n_a):
    y_list.append(hv.Curve((x,fct(a_var[i],x,b))).redim.range(x=(0,x_range[i]), y=(0,3)))
    
dict_y = {i:y_list[i] for i in range(n_a)}
hmap_y = hv.HoloMap(dict_y, kdims = ['a'])#.options(normalize=False, shared_axes=False)
hmap_y
```
