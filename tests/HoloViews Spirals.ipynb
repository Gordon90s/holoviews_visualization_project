```python
import holoviews as hv
from holoviews import Store
import numpy as np
hv.extension('matplotlib')
```

```python
%opts Path (linewidth=2 color=Palette('Blues'))
#%opts Path (linewidth=2 color='black')
%opts Path [fig_size=100]



def spiral_equation(f, ph2):
    r = np.arange(0, 1, 0.005)
    xs, ys = (r * fn(f*np.pi*np.sin(r)+ph2) for fn in (np.cos, np.sin))
    paths = hv.Path([(xs,ys)], extents=(-0.22,-0.22,0.22,0.22))
    return paths
```

```python
spiral_dict_2D ={(f,ph2:spiral_equation(f,ph2) for f in [1] for ph2 in [1,2,4,5,6,7,8]}
```
